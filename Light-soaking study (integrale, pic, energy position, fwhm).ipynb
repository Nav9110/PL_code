```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
%matplotlib inline
from numpy import trapz
from scipy.interpolate import interp1d
```

```python
# Chemin du répertoire + nom du fichier
chemin = r'C:\Users\navid.mouhamad\OneDrive - IPVF\Desktop\Manips\Campagne\Stage\20240926\Episodes_light_soaking.csv'
df = pd.read_csv(chemin, sep=';')
df
```

```python
x = df.iloc[:, 0]
y = df.iloc[:, 2]

interpolation = interp1d(x, y, kind='linear', fill_value='extrapolate')
# Borne d'intégration 
x_inter = np.linspace(700, 950, 1000)
y_inter = interpolation(x_inter)
aire = trapz(y_inter, x_inter)
print("Intégrale de la courbe interpolée :", aire)

plt.plot(x, y, '-', label='Données expérimentales')
plt.plot(x_inter, y_inter,'r-', label='Intégrale')
plt.fill_between(x_inter, y_inter, alpha=0.3)
plt.legend()
plt.grid()
plt.show()
```

```python
integrale = []
gap = []
fwhm_list = []

for i in range(1, len(df.columns), 1):
    y  = df.iloc[:, i] # intensity
    x = df.iloc[:, 0]  # wvl
    interpolation = interp1d(x, y, kind='linear', fill_value='extrapolate')
    x_inter = np.linspace(700, 950, 1000)
    y_inter = interpolation(x_inter)
    
    # Integrale
    aire = trapz(y_inter, x_inter)
    print("Intégrale de la courbe interpolée :", aire)
    integrale.append(aire)
    
    # Pic de PL
    val_gap = np.argmax(y_inter)
    x_max = x_inter[val_gap]
    print("Gap de la courbe :", x_max)
    gap.append(x_max)
    
    # FWHM
    y_max = np.max(y_inter)
    half_max = y_max / 2
    indices_above_half_max = np.where(y_inter >= half_max)[0]
    x_left = x_inter[indices_above_half_max[0]]
    x_right = x_inter[indices_above_half_max[-1]]
    fwhm = x_right - x_left
    print("FWHM :", fwhm)
    fwhm_list.append(fwhm)
    
    plt.plot(x, y,'-', label=df.columns[i])
    plt.plot(x_inter, y_inter,'r-', label='Intégrale')
    plt.fill_between(x_inter, y_inter, alpha=0.3)
    plt.legend()
    plt.show()
    
```

```python
# Conversion en eV
bandgap = []
for i in gap:
    Eg = 1240/i
    bandgap.append(round(Eg, 3))
Eg = bandgap
print(Eg)
```

```python
# Integrale vs time
x = np.arange(10,710,10)

plt.plot(x, integrale[1:], 'o', label='Données Intégrale')
plt.xlabel('Time (s)')
plt.ylabel('Integrated PL intensity (W/m²/sr)')
plt.grid()
plt.legend()
plt.show()
```

```python
# PL peak vs time
x = np.arange(10,710,10)

plt.plot(x, Eg[1:], 'o', label='Données PL pics')
plt.xlabel('Time (s)')
plt.ylabel('PL peak shift (eV)')
plt.grid()
plt.legend()
plt.show()
```

```python
# FWHM vs time
x = np.arange(10,710,10)

plt.plot(x, fwhm_list[1:], 'o', label='Données FWHM')
plt.xlabel('Time (s)')
plt.ylabel('FWHM shift (nm)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Sauvegarde dans un dataframe "Integrated_PL_gap"
Integrated_PL_peak_fwhm = pd.DataFrame()

Integrated_PL_peak_fwhm['Time (s)'] = x
Integrated_PL_peak_fwhm['Integrated PL Intensity (W/m²/sr)'] = integrale[1:]
Integrated_PL_peak_fwhm['PL peak (eV)'] = Eg[1:]
Integrated_PL_peak_fwhm['FWHM (nm)'] = fwhm_list[1:]
#Integrated_PL = fichier.dropna()

# Conversion en csv (dans le dossier ou se situe le code)
Integrated_PL_peak_fwhm.to_csv('Integrated_PL_peak_fwhm.csv', index=False)

Integrated_PL_peak_fwhm
```
